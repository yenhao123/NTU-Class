```python
from sklearn import datasets
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
in_path = Path("hw1_train.dat")

df = pd.read_csv(in_path, delimiter=' ', header=None)
df = df.drop(columns=[0])
df = df.drop([256])
n_data, n_dim = len(df), len(df.columns) - 1
print("Data size : {}; Data dimension : {}".format(n_data, n_dim))
```

```python
x_train, y_train = df.iloc[:, :12].to_numpy(), df.iloc[:, 12].astype(int).to_numpy()
```

```python
def sign(x):
    if x > 0:
        return 1
    else:
        return -1
```

```python
def is_data_correct(x_train, y_train, w):
    corect = 0
    for x, y in zip(x_train, y_train):
        if sign(np.dot(w, x)) != y:
            return False
    return True
```

```python
def random_PLA(x_train, y_train, seed=7):
    w = np.zeros(x_train.shape[1])
    n_iteration, n_correct = 0, 0
    np.random.seed(seed)
    while n_correct <= 5 * n_data:
        n_iteration += 1
        selected_idx = np.random.choice(x_train.shape[0], 1)[0]
        x, y = x_train[selected_idx], y_train[selected_idx]
        if sign(np.dot(w, x)) != y:
            w = w + y * x 
            print(n_iteration, w)
            n_correct = 0
        else:
            n_correct += 1
    
    #if is_data_correct:
    #    print("All data correct")
    
    return n_iteration
```

## Quiz 9

```python
# Concate bias
x_train_baseline = np.array([np.concatenate(([1], x)) for x in x_train])

# Run PLA
n_epoch = 1000
n_iterations = []
for epoch in range(n_epoch):
    n_iteration = random_PLA(x_train_baseline, y_train, seed=epoch)
    print("Epoch : {}; Numbers of iteration : {}".format(epoch+1, n_iteration))
    n_iterations.append(n_iteration)
```

```python
sns.histplot(n_iterations)
```

```python
np.median(np.array(n_iterations))
```
